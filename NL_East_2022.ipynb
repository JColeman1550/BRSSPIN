```python
%%capture
import numpy as np
import pandas as pd
import pybaseball as pyb
import matplotlib.pyplot as plt
```

```python
year = 2022

braves = pyb.schedule_and_record(year,'ATL')
mets = pyb.schedule_and_record(year,'NYM')
phillies = pyb.schedule_and_record(year,'PHI')
marlins = pyb.schedule_and_record(year,'MIA')
nationals = pyb.schedule_and_record(year,'WSN')
```

## **Wins, Win Percentages, & Games Back**

Gathering wins, win % and GB for each team in the division

```python
braves['Wins'] = np.where(braves['W/L']=='W',1,(np.where(braves['W/L']=='W-wo',
                                                         1,0))).cumsum()
braves['Losses'] = np.where(braves['W/L'] == 'L',1,(np.where(braves['W/L']=='L-wo',
                                                             1,0))).cumsum()
braves['Win_Percentage'] = braves['Wins']/(braves['Wins'] + braves['Losses'])


mets['Wins'] = np.where(mets['W/L']=='W',1,(np.where(mets['W/L']=='W-wo',
                                                     1,0))).cumsum()
mets['Losses'] = np.where(mets['W/L']=='L',1,(np.where(mets['W/L']=='L-wo',
                                                      1,0))).cumsum()
mets['Win_Percentage'] = mets['Wins']/(mets['Wins'] + mets['Losses'])


phillies['Wins'] = np.where(phillies['W/L']=='W',1,(np.where(phillies['W/L']=='W-wo',
                                                             1,0))).cumsum()
phillies['Losses'] = np.where(phillies['W/L']=='L',1,(np.where(phillies['W/L']=='L-wo',
                                                               1,0))).cumsum()
phillies['Win_Percentage'] = phillies['Wins']/(phillies['Wins'] + phillies['Losses'])


marlins['Wins'] = np.where(marlins['W/L']=='W',1,(np.where(marlins['W/L']=='W-wo',
                                                           1,0))).cumsum()
marlins['Losses'] = np.where(marlins['W/L']=='L',1,(np.where(marlins['W/L']=='L-wo',
                                                             1,0))).cumsum()
marlins['Win_Percentage'] = marlins['Wins']/(marlins['Wins'] + marlins['Losses'])                                                                                                                        


nationals['Wins'] = np.where(nationals['W/L']=='W',1,(np.where(nationals['W/L']=='W-wo',
                                                               1,0))).cumsum()
nationals['Losses'] = np.where(nationals['W/L']=='L',1,(np.where(nationals['W/L']=='L-wo',
                                                                 1,0))).cumsum()
nationals['Win_Percentage'] = nationals['Wins']/(nationals['Losses'])                                                                                                                                
```

# **Graphing Wins & Win Percentages**

We see that the division race came down to a tigjhtly contested race between the Braves and Mets. Both teams finished the regualr season with 101 wins (Braves ultimately took the division winning  the tiebreaker game 5-3). At about the 125 game mark the Braves caught up to the Mets after trailing them in the standings all season long. 

```python
plt.rcParams['figure.figsize'] = (8,6)

plt.plot(braves['Wins'],label='Braves',c='navy')
plt.plot(mets['Wins'],label='Mets',c='darkorange')
plt.plot(phillies['Wins'],label='Phillies',c='Red')
plt.plot(marlins['Wins'],label='Marlins',c='cyan')
plt.plot(nationals['Wins'],label='Nationals',c='darkred')

plt.xticks(np.arange(0,len(braves.index),step=10)) 

plt.xlabel('Game')
plt.ylabel('Wins')
plt.grid(True)
plt.legend(loc='lower right')

plt.title("NL East Wins Comparison({})".format(year))
```

```python
plt.rcParams['figure.figsize']=(8,6)

plt.xticks(np.arange(0,len(braves.index),step=10))

plt.plot(braves['Win_Percentage'],label='Braves',c='navy')
plt.plot(mets['Win_Percentage'],label='Mets',c='darkorange')
plt.plot(phillies['Win_Percentage'],label='Phillies',c='red')
plt.plot(marlins['Win_Percentage'],label='Marlins',c='cyan')
plt.plot(nationals['Win_Percentage'],label='Nationals',c='maroon')

plt.xlabel('Game')
plt.ylabel('Win Percentage')
plt.legend(loc='lower right')
plt.grid(True)

plt.title('NL East Wins Percentage({})'.format(year))
```

# **Graphing Games Back**

Creating a column to numerically represent GB

```python
#Graphing Games Back

def modify_gb(gb):
  if gb == 'Tied':
    gb = 0.0
  elif 'up' in gb:
    gb = 0.0
  else:
    gb = -float(gb)
  
  return(gb)
```

```python
braves['int_GB'] = braves['GB'].dropna().apply(modify_gb)
mets['int_GB'] = mets['GB'].dropna().apply(modify_gb)
phillies['int_GB'] = phillies['GB'].dropna().apply(modify_gb)
marlins['int_GB'] = marlins['GB'].dropna().apply(modify_gb)
nationals['int_GB'] = nationals['GB'].dropna().apply(modify_gb)
```

# **Games Back Graph**

The Nationals were the first team in the division to fall 10 games back at the 30 game mark. They finished the season 46 games back. 

At the midseason point the divison took its shape as it became a clear two team race between the Braves and Mets. The Phillies hung in and got within 7 games at about the 85 game mark but ended up finishing 14 games back from first. 

The Mets had a stronghold on first place for much of the season even obataing a season high 11 game lead at the 50 game mark. Between game 50 and 95 the Mets sqaandered the lead allowing the Braves to get within 1 game. The Braves then fell off a bit losing 6 games between games 95 and 111. Pulling a miracle comeback the Braves closed the divison gap for good at game 158 where they would go on to beat the Mets in a tiebreaker to become NL East Champions. 

```python
plt.rcParams['figure.figsize']=(8,6)

plt.plot(braves['int_GB'],label='Braves',c='navy')
plt.plot(mets['int_GB'],label='Mets',c='darkorange')
plt.plot(phillies['int_GB'],label='Phillies',c='red')
plt.plot(marlins['int_GB'],label='Marlins',c='cyan')
plt.plot(nationals['int_GB'],label='Nationals',c='maroon')

plt.xticks(np.arange(0,len(braves.index),step=10))
plt.yticks(np.arange(0, -50, step=-5))
plt.xlabel('Game')
plt.ylabel('Games Back (in negative)')
plt.grid(True)

plt.title("NL East Games Back Comparison ({})".format(year))
plt.legend(loc='lower left')
```

# Graphing Total Runs & Runs Allowed

creating new columns for Total Runs & RA

```python
#Graphing Total Runs & Runs Allowed

braves['Total_Runs'] = braves['R'].cumsum()
braves['Total_RA'] = braves['RA'].cumsum()

mets['Total_Runs'] = mets['R'].cumsum()
mets['Total_RA'] = mets['RA'].cumsum()

phillies['Total_Runs'] = phillies['R'].cumsum()
phillies['Total_RA'] = phillies['RA'].cumsum()

marlins['Total_Runs'] = marlins['R'].cumsum()
marlins['Total_RA'] = marlins['RA'].cumsum()

nationals['Total_Runs'] = nationals['R'].cumsum()
nationals['Total_RA'] = nationals['RA'].cumsum()

```

```python
plt.rcParams['figure.figsize']=(16,8)

#Total Runs
plt.subplot(1,2,2)
plt.plot(braves['Total_Runs'],label='Braves',c='navy')
plt.plot(mets['Total_Runs'],label='Mets',c='darkorange')
plt.plot(phillies['Total_Runs'],label='Phillies',c='red')
plt.plot(marlins['Total_Runs'],label='Marlins',c='cyan')
plt.plot(nationals['Total_Runs'],label='Nationals',c='maroon')

plt.xticks(np.arange(0,len(braves.index),step=10))
plt.xlabel('Game')
plt.ylabel('Total Runs')
plt.grid(True)
plt.legend(loc='lower right')
plt.title("NL East Total Runs ({})".format(year))


#Total Runs Allowed

plt.subplot(1,2,1)
plt.plot(braves['Total_RA'],label='Braves',c='navy')
plt.plot(mets['Total_RA'],label='Mets',c='darkorange')
plt.plot(phillies['Total_RA'],label='Phillies',c='red')
plt.plot(marlins['Total_RA'],label='Marlins',c='cyan')
plt.plot(nationals['Total_RA'],label='Nationals',c='maroon')

plt.xticks(np.arange(0,len(braves.index),step=10))
plt.xlabel('Game')
plt.ylabel('Total Runs Allowed')
plt.legend(loc='lower right')
plt.grid(True)
plt.title("NL East Total Runs Allowed ({})".format(year))
```

The Marlins are an interesting case. While giving up less than 700 runs (good for 3rd in the division) they were dead last in the division in runs scored at just under 600. No other team was under 600 total runs scored. 

The Phillies scored at comparable rates with the Braves and Mets but their pitching and defense didn't hold up as well allowing the second most runs in the division (only in front of the last place Nationals).

Again the Braves and Mets led the way in Runs Scored and Runs Allowed. The Braves took the lead in Total Runs at the 115 game mark and held on to it for the rest of the season. It took until the last regular season game for the Mets to take the lead in Total Runs Allowed. 

Based on both graphs predicting the Braves to win the divison would be the right pick. Not only did they outscore the rest of the division for most of the second half but they also kept their Runs Allowed close to the Mets all season long.
