```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stadium = pd.read_csv('https://raw.githubusercontent.com/jldbc/pybaseball/master/pybaseball/data/mlbstadiums.csv')

stadium['y'] = stadium['y'] * -1
stadium = stadium.loc[:,'team':]

stadium
```

```python
def plot_stadium(team, color):

    team_df = stadium[stadium['team'] == team.lower()]
    for i in stadium['segment'].unique():
        data = team_df[team_df['segment'] == i]
        plt.plot(data['x'],data['y'], linestyle = '-', color = color)
    plt.suptitle(team.capitalize(), y=1.01, fontsize=15)
    plt.title(team_df['location'].any(), fontsize=10)
    plt.axis('on');
```

```python
pyb.playerid_lookup('trevor', 'story')
```

```python

start_date = '2021-01-01'
end_date = '2021-12-28'

df = pyb.statcast_batter(start_date, end_date, 596115)

home_df = df.loc[df['home_team'] == 'COL']

HR_COL = home_df[home_df['events'] == 'home_run']
```

```python
plot_stadium('rockies','purple')
plot_stadium('red_sox','red')

plt.plot(HR_COL['hc_x'],HR_COL['hc_y']*-1, 'o',color= 'black')

plt.suptitle(None)
plt.title('2021 Rockies (Purple) and Red Sox (Red)',fontsize = 15);
```
